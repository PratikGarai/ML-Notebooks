```python
class Perceptron :
    def __init__(self, input_dim, output_dim, lr, classes):
        self.inp = input_dim
        self.out = output_dim
        self.w = np.random.rand(self.inp,self.out)
        self.b = np.random.rand(self.out)
        self.lr = lr
        self.c = classes
        
    def sigmoid(self, x):
        return 1/(1+np.exp(-x))
    
    def sigmoid_derivative(self,x):
        return self.sigmoid(x)*(1-self.sigmoid(x))
    
    def train(self, x, y):
        mul = np.apply_along_axis(lambda i : np.dot(i,self.w)+self.b , 1, x) 
        sig = self.sigmoid(mul)*self.c
        err = (sig-y)**2
        
        loss = err.sum()
        
        dE_da = sig-y
        da_do = self.sigmoid_derivative(mul)
        dE_dw = np.dot(x.T, dE_da*da_do)
        
        self.w = self.w - self.lr*dE_dw 
        for i in dE_da*da_do:
            self.b -= i
        return loss
    
    def fit(self, x, y, epochs=1000, printCost = False, printGap = 100):
        for i in range(epochs):
            c = self.train(x,y)
            if printCost and i%printGap==0:
                print("Epoch",i,"\t Cost: ",c)
    
    def predict(self, x):
        return self.sigmoid(np.dot(x,self.w)+self.b)
```

```python
p = Perceptron(4,1, 0.01, 1)
p.fit(x,y,10000,True,1000)
```

```python
def conv(x):
    if x<0.5:
        return 0
    return 1
```

```python
# Random check
n = 20
nums = np.random.randint(0, 100, 20)
ls = []
for i in sorted(nums):
    ls.append([i, conv(p.predict(x[i])), y[i][0]])

d = pd.DataFrame(ls, columns=['index', 'predicted', 'actual'])
d
```

```python
# Full check
c = 0
n = 100
for i in range(n):
    if conv(p.predict(x[i]))==y[i][0]:
        c += 1

print("Accuracy : ",c/n)
```
